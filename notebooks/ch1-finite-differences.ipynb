# Finite Difference Demo: Second Derivative Approximation

This notebook demonstrates the second derivative finite difference approximation and shows how round-off error affects accuracy for very small step sizes.

```python
# Compute the error for a series of different h-values
hs = np.logspace(-20, 0, 21)

# List to collect the errors
errors = []

# Point in which we are interested
x0 = 1
f2 = []
# Create domain and function values for each domain
for h in hs:
    dyy = (1/h**2)*(np.exp(x0+h)+np.exp(x0-h)- 2*np.exp(x0))
    f2.append(dyy)
    
plt.semilogx(hs, f2, marker='o', color='b', label='C.1')
plt.legend()
plt.ylabel('f"')
plt.xlabel('$h$ Discretization size')
plt.title("C.1")
plt.grid(alpha=0.3)
```

```python
# Compute the error for a series of different h-values
hs = np.logspace(-20, 0, 21)

# List to collect the errors
errors = []

# Point in which we are interested
x0 = 1
# Create domain and function values for each domain
for h in hs:
    dyy = (1/h**2)*(np.exp(x0+h)+np.exp(x0-h)- 2*np.exp(x0))
    # Compute error
    error = np.max(np.abs(dyy - np.exp(x0))) 
    # Store the result
    errors.append(error)
    
# Create an ndarray from the list.
errors = np.asarray(errors)
# Plot the error on a log-log plot
plt.semilogx(hs, errors, marker='o', color='b', label='Absolute error')
# plot the linear line
hs = np.logspace(-12, 0, 16)
# plt.loglog(hs, 0.5 * np.sin(x0) * hs, color='r', ls='--')
plt.loglog(hs, hs/6, linewidth=2.0, label=r'$E(h) = \frac{h}{6}$')
plt.loglog(hs, (hs**2)/12, linewidth=2.0, label=r'$E(h) = \frac{h^2}{12}$')
plt.legend()
plt.ylabel('Absolute Error')
plt.xlabel('$h$ Discretization size')
plt.grid(alpha=0.3)
```
